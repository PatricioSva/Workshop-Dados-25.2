```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

temperaturas=np.random.uniform(20,35, size = 7)
temperaturas=np.round (temperaturas, 1)

dados= {
    'Dia': ['Segunda-Feira', 'Terça-Feira', 'Quarta-Feira', 'Quinta-Feira', 'Sexta-Feira', 'Sábado', 'Domingo'],
    'Temperatura (°C)': [ 25.6, 34.3, 31.0, 29.0, 22.3, 22.3, 20.9 ]
}

df= pd.DataFrame(dados)
media_semanal= df['Temperatura (°C)'].mean()

print(df)
print(f'\nMédia da semana: {media_semanal:.2f} °C')

plt.figure(figsize=(10,5))
plt.plot(df['Dia'], df['Temperatura (°C)'], marker= 'o', color = 'orange')
plt.axhline(media_semanal,linestyle='--', color='orange')
plt.title ('Variação da Temperatura ao longo da semana')
plt.xlabel('Dia')
plt.ylabel('Temperatura (°C)')
plt.grid (True)
plt.tight_layout()
plt.show()

```

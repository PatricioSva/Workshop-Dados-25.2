```python
import pandas as pd

# Lendo o dataset
df = pd.read_csv('arquivo.csv')

# Garantindo que a coluna 'churn' existe (case-insensitive)
col_churn = [col for col in df.columns if 'churn' in col.lower()][0]

# Padronizando os valores para minúsculo (using the identified column name)
df[col_churn] = df[col_churn].str.strip().str.lower()

# Contagem dos valores
contagem = df[col_churn].value_counts()

# Cálculo das porcentagens (using the identified column name and lowercased values)
percent_yes = (contagem.get('yes', 0) / len(df)) * 100
percent_no = (contagem.get('no', 0) / len(df)) * 100

# Exibindo resultados
print(f"Taxa de Churn (cancelaram): {percent_yes:.2f}%")
print(f"Taxa de Retenção (não cancelaram): {percent_no:.2f}%")
```
